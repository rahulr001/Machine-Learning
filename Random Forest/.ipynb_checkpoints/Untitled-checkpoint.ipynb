```python
from sklearn.datasets import load_iris
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
import pandas as pd
```

```python
import numpy as np
```

```python
np.random.seed(0)
```

```python
irish  = load_iris()
```

```python
df = pd.DataFrame(irish.data, columns=irish.feature_names)
```

```python
df.head()
```

```python
len(df)
```

```python
df['species'] = pd.Categorical.from_codes(irish.target, irish.target_names)
```

```python
df.head()
```

```python
irish
```

```python
df['is_train'] = np.random.uniform(0, 1, len(df)) <= .75
```

```python
df.head()
```

```python
train, test = df[df['is_train'] == True], df[df['is_train']==False]
```

```python
train
```

```python
test
```

```python
features = df.columns[:4]
```

```python
features
```

```python
Y = pd.factorize(train['species'])[0]
```

```python
Y
```

```python
clf = RandomForestClassifier(n_jobs=2,random_state=0)
```

```python
clf.fit(train[features], Y)
```

```python
clf.predict(test[features])
```

```python
clf.predict_proba(test[features])
```

```python
preds = irish.target_names[clf.predict(test[features])]
```

```python
preds
```

```python
pd.crosstab(test.species, preds, rownames=['actual'], colnames=['predited'])
```

```python
preds = irish.target_names[clf.predict([[1,2,3,4]])]
```
