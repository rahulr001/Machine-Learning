```python
import matplotlib.pyplot as plt
```

```python
from sklearn import linear_model
```

```python
df = pd.read_csv('example-1.csv')
```

```python
df
```

```python
plt.xlabel('year')
plt.ylabel('price')
plt.scatter(df['year'], df['per capita income (US$)'], color = "red", marker="+")
```
